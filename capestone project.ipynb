```python
from statsmodels.tsa.stattools import adfuller
```

```python
result=adfuller(data1['Weekly_Sales'])
pvalue=result[1]
```

```python
pvalue
```

```python
rolling_mean = data1.rolling(window = 52).mean()
rolling_mean_detrended = data1 - rolling_mean
```

```python
rolling_mean_detrended[:53]
```

```python
ax1=plt.subplot(1,2,2)
rolling_mean_detrended.plot(figsize=(12,4),color='tab:red',
                            title='differenced with rolling mean',
                            ax=ax1)

ax2=plt.subplot(1,2,1)
data.plot(figsize=(12,4),
          color='tab:red',
          title="original data",
          ax=ax2)

plt.show()
```

```python
rolling_mean_detrended[:53]
```

```python
rolling_mean_detrended.shift()[:53]
```

```python
rolling_mean_detrended
```

```python
rolling_mean_detrended.shift()
```

```python
rolling_mean_detrended_diff = rolling_mean_detrended - rolling_mean_detrended.shift()
```

```python
ax1=plt.subplot(1,2,2)
rolling_mean_detrended_diff.plot(figsize=(12,4),color='tab:red',
                            title='differenced timeseries data',
                            ax=ax1)

ax2=plt.subplot(1,2,1)
data.plot(figsize=(12,4),
          color='tab:red',
          title="original data",
          ax=ax2)

plt.show()
```

```python
rolling_mean_detrended_diff = rolling_mean_detrended_diff.dropna()
```

```python
rolling_mean_detrended_diff
```

```python
result = adfuller(rolling_mean_detrended_diff['Weekly_Sales'])
pvalue = result[1]
```

```python
pvalue
```

```python
ax1=plt.subplot(1,2,2)
rolling_mean_detrended.plot(figsize=(12,4),color='tab:red',
                            title='new data',
                            ax=ax1)

ax2=plt.subplot(1,2,1)
rolling_mean_detrended_diff.plot(figsize=(12,4),color='tab:red',
                            title='differenced timeseries data',
                            ax=ax2)
```

```python
train = rolling_mean_detrended_diff.iloc[:70]['Weekly_Sales']
test = rolling_mean_detrended_diff.iloc[70:]['Weekly_Sales']
```

```python
train
```

```python
test
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX,SARIMAXResults
```

```python
model = SARIMAX(train,order = (1,0,1),seasonal_order=(1,0,1,52))
model = model.fit()
```

```python
len(train)
```

```python
len(test)
```

```python
rolling_mean_detrended_diff['predict']=model.predict(start=len(train),end=len(train)+len(test)-1,dynamic=True)
```

```python
rolling_mean_detrended_diff[['Weekly_Sales','predict']].plot()
```

```python
forecast = model.forecast(steps = 100)
rolling_mean_detrended_diff.plot()
forecast.plot()
```

```python
forecast=model.forecast(step=60)
rolling_mean_detrended_diff.plot()
forecast.plot()
```
